```python
WT="/mnt/sunxh/arabidopis_dataset/predict.col.feature.tsv"
KO="/mnt/sunxh/arabidopis_dataset/predict.vir.feature.tsv"
WT2 = "/mnt/sunxh/arabidopis_dataset/predict.virc.feature.tsv"

files=[WT,KO,WT2]

def process_file(file):
    mod_dict = {}
    with open(file) as f:
        for line in f:
            line = line.rstrip()
            id = "|".join(line.split("\t")[0:2])
            probability = float(line.split("\t")[5])
            if id not in mod_dict:
                mod_dict[id] = [0, 0]
            if probability > 0.95:
                mod_dict[id][0] += 1
            mod_dict[id][1] += 1
    return mod_dict

mod_dicts = []

for file in files:
    mod_dict = process_file(file)
    mod_dicts.append(mod_dict)
```

```python
import traceback

from plotnine import *
import pandas as pd
import re
import numpy as np
id_dict={}
def build_id_dict(file):
    id_dict = {}
    with open(file) as f:
        for line in f:
            if ">" in line:
 
                id = line.split(" ")[0].split(".")[0][1:]
                
 
                if "gene_symbol:" in line:
                    gene_name = line.split("gene_symbol:")[1].split(" ")[0]
 
                elif "gene:" in line:
                    gene_name = line.split("gene:")[1].split(" ")[0]
 
                else:
                    gene_name = id
                
                id_dict[id] = gene_name
                
    return id_dict


id_dict=build_id_dict("/mnt/sunxh/arabidopis_dataset/Arabidopsis_thaliana.TAIR10.cdna.all.fa")
```
